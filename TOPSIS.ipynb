```python
#导入分析库
import pandas as pd
import numpy as np
from pandas import Series,DataFrame
 
##导入数据
df = pd.read_excel('C:/Users/马嘉文/Desktop/5910project/clean_data/demo.xls')
label_need=df.keys()[1:]
data1=df[label_need].values
data2=data1
index1=[1] 
a=120    #最优型数值
for i in range(0,len(index1)):
    data2[:,index1[i]]=1-abs(data1[:,index1[i]]-a)/max(abs(data1[:,index1[i]]-a))
print(data2)
```

```python
from sklearn.preprocessing import scale
import numpy as np

data2_scale=scale(X=data2,with_mean=True,with_std=True,copy=True)
print('标准化数据：\n',data2_scale)
```

```python
#计算组合权重
a=[5,2.75,3,4]
b=[0.21754368,0.04045323,0.39580811,0.34619498]
w=b
sum=0
for i in range(4):
    sum = sum + np.sqrt(a[i]*b[i])
#     print(sum)
for i in range(4):
    w[i] = np.sqrt(a[i]*b[i])/sum
print(w)
```

```python
#得到加权后的数据
R=data2*w
print(R)
```

```python
#得到最大值最小值距离
r_max=np.max(R, axis=0)  #每个指标的最大值
r_min=np.min(R,axis=0)   #每个指标的最小值
d_z = np.sqrt(np.sum(np.square((R -np.tile(r_max,(250,1)))),axis=1))  #d+向量
d_f = np.sqrt(np.sum(np.square((R -np.tile(r_min,(250,1)))),axis=1))  #d-向量 
print('每个指标的最大值',r_max)
print('每个指标的最小值',r_min)
print('d+向量',d_z)
print('d-向量',d_f)
#d_z1 = DataFrame(d_z,columns=['score'
#,'duration','comments','rating_num'
#])
#d_z1.to_excel('C:/Users/马嘉文/Desktop/test1.xlsx', sheet_name='Sheet1', na_rep='', float_format=None, columns=None, header=True, index=True, index_label=None, startrow=0, startcol=0, engine=None, merge_cells=True, encoding=None, inf_rep='inf', verbose=True, freeze_panes=None)
```

```python
#得到评分
s=d_f/(d_z+d_f )
Score=100*s/max(s)
for i in range(0,len(Score)):
    print(Score[i])
```
