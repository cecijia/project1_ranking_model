```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt

df=pd.read_csv("C:/Users/马嘉文/Desktop/5910project/original_data/top250_data.csv")
print(df.head())
```

```python
df.describe()
```

```python
plt.figure(figsize=(20 ,10 ))
plt.xticks(rotation=45)
sns.countplot(data = df,x=df['year'])
plt.show()
```

```python
plt.figure(figsize=(10 ,7))
plt.xticks(rotation=45)
sns.countplot(data = df,y=df['score'])
plt.show()
```

```python
#如遇中文显示问题可加入以下代码
from pylab import mpl
mpl.rcParams['font.sans-serif'] = ['SimHei'] # 指定默认字体
mpl.rcParams['axes.unicode_minus'] = False # 解决保存图像是负号'-'显示为方块的问题
a,ax=plt.subplots(figsize=(6,4))
sns.countplot(data=df, x="type1",palette="Greens_r")
plt.show()
```

```python
a,ax=plt.subplots(figsize=(15,10))
sns.countplot(data=df, y="score", hue="type1",palette='Paired_r')
plt.show()
```

```python
sns.jointplot(data=df, x="year", y="score")
plt.show()
```
