```python
import copy
import pandas as pd
import numpy as np
from pandas import Series,DataFrame

#读取数据
data=pd.read_excel('C:/Users/马嘉文/Desktop/demo.xls')
#print(data)
label_need=data.keys()[1:]
#print(label_need)
#data=data(2:end) #%只取指标数据
data1=data[label_need].values
print(data1)
```

```python
#正向化
#指标正向    化处理后数据为data2
data2=data1
print(data2)
```

```python
#正向化
#某点最优型指标
index1=[1] 
a=120    #最优型数值
for i in range(0,len(index1)):
    data2[:,index1[i]]=1-abs(data1[:,index1[i]]-a)/max(abs(data1[:,index1[i]]-a))
print(data2)
```

```python
#0.002~1区间归一化
[m,n]=data2.shape
data3=copy.deepcopy(data2)
ymin=0.002
ymax=1
for j in range(0,n):
    d_max=max(data2[:,j])
    d_min=min(data2[:,j])
    data3[:,j]=(ymax-ymin)*(data2[:,j]-d_min)/(d_max-d_min)+ymin
#print(data3)
data4 = DataFrame(data3,columns=['score','duration','comments','rating_num'])
data4.to_excel('C:/Users/马嘉文/Desktop/test2.xlsx', sheet_name='Sheet1', na_rep='', float_format=None, columns=None, header=True, index=True, index_label=None, startrow=0, startcol=0, engine=None, merge_cells=True, encoding=None, inf_rep='inf', verbose=True, freeze_panes=None)
```

```python
#计算信息熵
p=copy.deepcopy(data3)
for j in range(0,n):
    p[:,j]=data3[:,j]/sum(data3[:,j])
print(p)
E=copy.deepcopy(data3[0,:])
for j in range(0,n):
    E[j]=-1/np.log(m)*sum(p[:,j]*np.log(p[:,j]))
print(E)
```

```python
# 计算权重
w=(1-E)/sum(1-E)
print(w)
```

```python
#计算得分
s=np.dot(data3,w)
Score=100*s/max(s)
for i in range(0,len(Score)):
    print(f"第{i}个评价对象得分为：{Score[i]}")
```
